```python
class Income:
        
    def get_total_income(self):
        self.total_income = 0
        self.rent = int(input("How much do you expect to make every month in rent? "))
        self.total_income += self.rent
        self.add_ons = input("Is there additional income from your rental property such as laundry, storage, etc.?\nEnter Y or N: ")
        if self.add_ons.lower() == 'y':
            self.additional = int(input("How much additional income will you make? "))
            self.total_income += self.additional
        print('='*100)
        print(f"Total Monthly Income:\n\t${self.total_income}")
        print('='*100)
```

```python
class Expenses:
    
    def get_expenses(self):
        self.total_expenses = 0
        self.taxes = int(input("Taxes are the worst, but until we can topple the government we have to pay them so\nwe'll get them out of the way first. How much will you pay in taxes every month for this home? "))
        self.total_expenses += self.taxes
        self.insurance = int(input("Insurance is almost as bad as taxes. How much is that going to run you every month? "))
        self.total_expenses += self.insurance
        utilities = input("Do you plan on paying for utilities such as electric, water, sewer, etc. yourself (Y/N)? ")
        if utilities.lower() == 'y':
            self.utilities = int(input("How much will you be paying monthly for utilities? "))
            self.total_expenses += self.utilities
        elif utilities.lower() != 'n':
            utilities = input("That is an invalid selection, please enter Y or N: ")
            self.utilities = int(input("How much will you be paying monthly for utilities? "))
            self.total_expenses += self.utilities
        vacant = input("Will you be setting aside money for a vacancy fund (Y/N)? ")
        if vacant.lower() == 'y':
            self.vacancy = int(input("How much will you be setting aside for that fund? "))
            self.total_expenses += self.vacancy
        elif vacant.lower() != 'n':
            vacant = input("That is an invalid selection, please enter Y or N: ")
            self.vacancy = int(input("How much will you be setting aside for that fund? "))
            self.total_expenses += self.vacancy
        self.repairs = int(input("We all know that nothing last forever.\nWith that in mind, how much do you plan on spending on repairs or replacements every month? "))
        self.total_expenses += self.repairs
        management = input("Will you be hiring a property manager (Y/N)? ")
        if management.lower() == 'y':
            self.prop_mgmt = int(input("What is the monthly cost of the property manager? "))
            self.total_expenses += self.prop_mgmt
        elif management.lower() != 'n':
            management = input("That is an invalid selection, please enter Y or N: ")
            self.prop_mgmt = int(input("What is the monthly cost of the porperty manager? "))
            self.total_expenses += self.prop_mgmt
        mortgage = input("Will you have a mortgage on this home (Y/N)? ")
        if mortgage.lower() == 'y':
            self.mortgage = int(input("How much is the monthly payment for the mortgage going to be? "))
            self.total_expenses += self.mortgage
        elif mortgage.lower() != 'n':
            mortgage = input("That is an invalid selection, please enter Y or N: ")
            self.mortgage = int(input("How much will you be setting aside for that fund? "))
            self.total_expenses += self.mortgage
        additional = input("Will there be any additional expeses such as HOA fees, lawn service, etc. (Y/N)? ")
        if additional.lower() == 'y':
            self.additional = int(input("What is the total monthly amount of the additional expenses? "))
            self.total_expenses += self.additional
        elif additional.lower() != 'n':
            additional = input("That is an invalid selection, please enter Y or N: ")
            self.additional = int(input("How much will you be setting aside for that fund? "))
            self.total_expenses += self.additional
        print('='*100)
        print(f"Total Monthly Expenses:\n\t${self.total_expenses}")
        print('='*100)
            
        
```

```python
class CashOnCash:
    
    
    def cash_on_cash(self):
        self.total_investment = 0
        print("The last thing I need is your cash on cash information so I can figure our your expected ROI.")
        down_pay = input("Did you, or will you, make a downpayment on this home (Y/N)? ")
        if down_pay.lower() == 'y':
            self.down_pay = int(input("How much did you pay as a down payment? "))
            self.total_investment += self.down_pay
        elif down_pay.lower() != 'n':
            down_pay = input("That is an invalid selection, please enter Y or N: ")
            self.down_pay = int(input("How much did you pay as a down payment? "))
            self.total_investment += self.down_pay
        closing = input("Did you, or will you, pay any closing costs (Y/N)? ")
        if closing.lower() == 'y':
            self.closing = int(input("How much are closing costs? "))
            self.total_investment += self.closing
        elif closing.lower() != 'n':
            closing = input("That is an invalid selection, please enter Y or N: ")
            self.closing = int(input("How much are closing costs? "))
            self.total_investment += self.closing
        rehab = input("Do you have a budget for rehab after purchase (Y/N)? ")
        if rehab.lower() == 'y':
            self.rehab = int(input("How much is in that budget? "))
            self.total_investment += self.rehab
        elif rehab.lower() != 'n':
            rehab = input("That is an invalid selection, please enter Y or N: ")
            self.rahab = int(input("How much is that budget? "))
            self.total_investment += self.rehab
        misc = input("Did you have any other miscelaneous investement expenses (Y/N)? ")
        if misc.lower() == 'y':
            self.misc = int(input("How much are those expenses? "))
            self.total_investment += self.misc
        elif misc.lower() != 'n':
            misc = input("That is an invalid selection, please enter Y or N: ")
            self.misc = int(input("How much are those expenses? "))
            self.total_investment += self.misc
        print('='*100)
        print(f"Total Investment:\n\t${self.total_investment}")
        print('='*100)
        
```

```python
def get_roi():
    
    my_income = Income()
    my_expenses = Expenses()
    my_cashoncash = CashOnCash()
    
    
    while True:
        name = input("Welcome to your ROI calculator!\nLet's start off by getting your name: ")
        print(f"Thanks, {name.title()}!\nLet's get started by gathering some income information")
        my_income.get_total_income()
        print(f"Great job, {name.title()}! Now that income is figured, let's take a look at expenses.")
        my_expenses.get_expenses()
        monthly_cash_flow = my_income.total_income - my_expenses.total_expenses
        annual_cash_flow = monthly_cash_flow * 12
        my_cashoncash.cash_on_cash()
        my_roi = (annual_cash_flow / my_cashoncash.total_investment) * 100
        print('='*100)
        print("Now that I have all of the information I needed you can find your expected ROI for this home below:")
        print('='*100)
        print(f"""
        Total Monthly Income:      ${my_income.total_income:.2f}
        Total Monthly Expenses:    ${my_expenses.total_expenses:.2f}
        Monthly Cash Flow:         ${monthly_cash_flow:.2f}
        Annual Cash Flow:          ${annual_cash_flow:.2f}
        
        Return on Investment:      {my_roi:.2f}%
        """)
        break
        
```
